##   Web Scraping  Hockey_Teams_Data  in Python

```python
from bs4 import BeautifulSoup  # BeautifulSoup (for parsing HTML)
import pandas as pd 
import requests                # requests (for making HTTP requests)
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'  # target URL to scrape data.
```

```python
page = requests.get(url)                             # Use requests to fetch the page content at that URL.
```

```python
soup = BeautifulSoup(page.text, 'html')              # Parse the page’s HTML content using BeautifulSoup
```

```python
teams_names = soup.find_all('td', class_='name')     # Find all table cells with class 'name'
```

```python
teams_names                                          # raw html 
```

```python
team_list = []

for team in teams_names:
    name = team.get_text(strip=True)
    team_list.append(name)
```

```python
team_list
```

```python
team_df = pd.DataFrame(team_list, columns=['Team Name'])
```

```python
team_df.head()
```

```python
# Save the DataFrame to a CSV file
team_df.to_csv('Hockey_TeamNames_Scrapper.csv', index=False)
```
